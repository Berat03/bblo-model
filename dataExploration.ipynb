```python
df = pd.read_csv('./data/cleanedData.csv')
```

```python
print(df.isnull().sum())
print(df.isna().sum())
```

```python
# Select rows with at least one NaN/null value
rows_with_nan = df[df.isnull().any(axis=1)]

rows_with_nan.head()
# rows_with_nan.to_csv('./data/missingData.csv') 
# # No particular trend / events for missing data.
df = df.ffill()
print(df.isnull().sum())
print(df.isna().sum())
```

```python
df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])

# Convert Total to numeric (removing commas)
df['Total'] = df['Total'].str.replace(',', '').astype(int)

# Plot Datetime vs Total
plt.figure(figsize=(10, 6))
plt.plot(df['Datetime'], df['Total'], marker='o')
plt.title("Library Occupancy Over Time")
plt.xlabel("Datetime")
plt.ylabel("Total Occupancy")
plt.grid(True)
plt.show()
# Incorrect use of backfill

# Incorrect use of backfill

# Incorrect use of backfill

# Incorrect use of backfill
```
